## Problem 1: Percent Negative

Write a function that returns what percentage of a list is negative.

You can assume the list contains only numbers and is not empty.

```python


def percent_negative(numbers):
    count = 0
    for number in numbers:
        if number < 0:
            count += 1
    percent_neg = (count / len(numbers)) * 100
    return percent_neg
```

## Problem 2: Reversing & Removing BAD_VALUE in a List (Two Ways)

Write two functions that returns the reverse of a given list with all instances of "BAD_VALUE" removed.

The first function should use element-wise processing by list index. The second should use slicing within a list comprehension.

```python
bad_value = "BAD_VALUE"
def fixreverse_elem(list_elements):
    bad_value_removed_list = []
    for i in range(len(list_elements)-1, -1, -1):
        if list_elements[i] != bad_value:
            bad_value_removed_list.append(list_elements[i])
    return bad_value_removed_list
```

```python
# call your element-wise function with a couple of different lists and print the results here
print(fixreverse_elem([3, 1, 4, 1, "BAD_VALUE", 9]))
print(fixreverse_elem(["Base", "Foil", "Holographic", "Polychrome", "Negative"]))
print(fixreverse_elem([12, 4, "Mira", "BAD_VALUE", "Rumi", "Zoey", 18, "BAD_VALUE"]))
```

```python
def fixreverse_slc(list_elements):
    bad_value = "BAD_VALUE"
    return [element for element in list_elements[::-1] if element != bad_value]

```

```python
# call your slice + list comprehension function with a couple of different lists and print the results here
print(fixreverse_slc([3, 1, 4, 1, "BAD_VALUE", 9]))
print(fixreverse_slc(["Base", "Foil", "Holographic", "Polychrome", "Negative"]))
print(fixreverse_slc([12, 4, "Mira", "BAD_VALUE", "Rumi", "Zoey", 18, "BAD_VALUE"]))
```

## Problem 3: Combine List Pairwise Lexicographically

Write a function that combines two lists by pairwise taking the elements of each list, determining which comes first based on lexicographical order according to Python, e.g., ['a','b','c',4], [1,2,3,'d'] → [1,'a',2,'b',3,'c',4,'d'].

In case of two lists of different length, the elements of the longer list are appended at the end, e.g., [1,2,3,4],[10,20]=>[1,10,2,20,3,4].

```python
def combine_list(list1,list2):
    result = []

    len_short_list = 0
    longer_list = []
    if len(list1) > len(list2):
        len_short_list = len(list2)
        longer_list = list1
    else:
        len_short_list = len(list1)
        longer_list = list2

    for i in range(len_short_list):
        if str(list1[i]) < str(list2[i]):
            result.append(list1[i])
            result.append(list2[i])
        elif str(list2[i]) < str(list1[i]):
            result.append(list2[i])
            result.append(list1[i])
        else:
            result.append(list1[i])
            result.append(list2[i])
        
    for item in longer_list[len_short_list::]:
        result.append(item)
    return result


    
```

## Problem 4: Largest Digit

Write a function that returns the largest digit in a number, e.g., 783456 → 8. Assume the number is round (has no decimal places).

```python
def largest_digit(number):
    largest_number = 0
    number = str(abs(number))
    for digit in number:
        if int(digit) > largest_number:
            largest_number = int(digit)
    return largest_number
```

## Problem 5: Check Sorting in Three Ways

Write three functions that check if a list is sorted, returning True or False. The first function should use a for-loop, the second a while-loop, and the third should use recursion.

*Hint: A function may need more than one argument, but you should allow the function to be called with only one argument given.*

```python
def check_for(example_list):
    for i in range(len(example_list) - 1):
        if example_list[i] > example_list[i+1]:
            return False
    return True
    
def check_while(example_list):
    i = 0
    while i < len(example_list) -1:
        if example_list[i] > example_list[i+1]:
            return False
        i += 1
    return True

def check_recursive(example_list,i=0):
    if i >= len(example_list) -1:
        return True
    if example_list[i] > example_list[i+1]:
        return False
    return check_recursive(example_list, i+1)
```

## Problem 6: Filtering

Write a function `filter_to_numbers` that filters out elements from a list that are not numbers. Preserve the order of the list.

*Hint: Look up the `isinstance` function*

```python
def filter_to_numbers(listt):
    number_list = []
    for item in listt:
        if isinstance(item,(int,float)):
            number_list.append(item)
    return number_list
```

## Problem 7: Hofstadter Q-Sequence

Write a function that takes a number `n` and an empty array and then writes the first `n` [Hofstadter Q-Sequence](https://en.wikipedia.org/wiki/Hofstadter_sequence#Hofstadter_Q_sequence) into the empty array. Print the array. 

The sequence can be generated via a relation where $Q(n) = Q(n - Q(n-1)) + Q(n - Q(n-2))$ with $Q(1) = Q(2) = 1$. Note this sequence starts with $Q(1)$.

Thus, if you call `hofstadterq(5)`, it should thus produce `[1, 1, 2, 3, 3]`.

```python
def hofstadterq(sequence_length, sequence):
    if sequence_length <= 0:
        return sequence

    sequence.append(1)
    if sequence_length == 1:
        return sequence

    sequence.append(1)
    if sequence_length == 2:
        return sequence

    for index in range(2, sequence_length):
        previous_value = sequence[index - 1]
        second_previous_value = sequence[index - 2]

        next_value = (
            sequence[index - previous_value] +
            sequence[index - second_previous_value]
        )

        sequence.append(next_value)

    return sequence
```

# Problem 8: Screentime Moving Maximum

Write a function `max_k_window` that takes a list of hours spent viewing a screen every day and computes a moving maximum over every consecutive `k` days. Some days will be missing measurements and will instead have an empty string for a value.

**Caveats:**
* The days missing measurements should be included in your `k` day window, but should not affect maximum.
  * e.g.`max_k_window([8.7,'',9.2],3)` should return `[9.2]`
* If all days in a window are missing measurements then the maximum for that window should also have the value `None`.
  * e.g. `max_k_window(['', '', 5.3], 2)` should return `[None, 53]`
* Only compute the maximum as long as you have enough data points to fill the window.
  * e.g. `max_k_window([10.0,2.5,3.0,4.5,5.0,6.0,7.5,8.0],7)` should return a list of length 2.
  * e.g. `max_k_window([12.0,24.0,3.0,4.5,5.5],2)` should return a list of length 4.
  * e.g. `max_k_window([10.0,20.0,3.5], 4)` should return an empty list.

You can use these as example lists, but make sure that your algorithm works on other lists/windows.

```python
def max_k_window(screen_hours, window_size):
    results = []

    if window_size <= 0 or window_size > len(screen_hours):
        return results

    for start_index in range(len(screen_hours) - window_size + 1):
        current_max = None
        found_value = False

        for offset in range(window_size):
            value = screen_hours[start_index + offset]

            if value != "" and value is not None:
                if not found_value:
                    current_max = value
                    found_value = True
                elif value > current_max:
                    current_max = value

        if found_value:
            results.append(current_max)
        else:
            results.append(None)

    return results
```
