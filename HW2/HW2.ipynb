```python
def calculate_median(vectors):
    sorted_vectors = sorted(vectors)    #sort the function
    number_of_vectors = len(sorted_vectors)    #number of vectors in the list

    if number_of_vectors % 2 == 1:    #if statement to check if it is odd or not.
        median_value = sorted_vectors[number_of_vectors // 2]   # if number is odd, it is the middle value.
        return median_value
    else:
        mid = number_of_vectors // 2    # if number is even, then we have to sum of index of mid one below and above and add them and divide by 2 .
        median_value = (sorted_vectors[mid - 1] + sorted_vectors[mid]) / 2
        return median_value


print(f"""The median temperature for TaylorSville is {calculate_median(temperature_vector)} degrees Fahrenheit.
The median precipitation is {calculate_median(precipitation_vector)} inches.""")
```

### Task 1.4: Calculate the Variance and Standard Deviation of a Vector

Write a function that calculates and returns the [variance](https://en.wikipedia.org/wiki/Variance) of a vector. Pass the temperature and precipitation vectors into this function and print the result. Provide a written interpretation of your results. Is the variance high? Why, Why not? 

The variance is the average of the squared deviations from the mean, i.e.,

$$ Var(X) = {\sigma}^{2} = \frac{1}{N-1} \sum_{i=1}^{N} {{(x_i - \mu)}^2} $$

where $\mu$ is the mean of the vector. Hint: use your mean function to calculate it.

Finally, calculate the standard devation $\sigma$, defined as the square root of the variance. Explain which measure (variance or standard deviation) is more interpretable in this context, and explain why.

```python
def calculate_variance(vectors):
    mean = sum(vectors) / len(vectors)
    number_of_vectors = len(vectors)

    squared_deviations = 0
    for x in vectors:
        squared_deviations += (x - mean) ** 2
    variance = squared_deviations / (number_of_vectors - 1)
    return variance

def calculate_std(vectors):
    return math.sqrt(calculate_variance(vectors))


print("Temperature:")
print(f"Variance: {calculate_variance(temperature_vector):.5f}")
print(f"Standard Deviation: {calculate_std(temperature_vector):.5f}")
print("\nPrecipitation:")
print(f"Variance: {calculate_variance(precipitation_vector):.5f}")
print(f"Standard Deviation: {calculate_std(precipitation_vector):.5f}")
```

```python
def calculate_variance_np(vectors):
    variance = np.var(vectors, ddof=1) #np.var(array,ddof=1) for sample variance (n-1)
    return variance

def calculate_std_np(vectors):
    return np.std(vectors, ddof=1)  #np.std(array,ddof=1) for sample std (n-1)

print("Temperature:")
print(f"Variance: {calculate_variance_np(temperature_vector):.5f}")
print(f"Standard Deviation: {calculate_std_np(temperature_vector):.5f}")
print("\nPrecipitation:")
print(f"Variance: {calculate_variance_np(precipitation_vector):.5f}")
print(f"Standard Deviation: {calculate_std_np(precipitation_vector):.5f}")
```

The function should return an array containing two arrays. The first should be the counts for each bin, the second should contain the borders of the bins.

For `b=5` for the temperature vector, your output should look like this: 

`[[19, 102, 78, 72, 95], [11.0, 23.0, 35.0, 47.0, 59.0, 71.0]]`

Here, the first array gives the size of these bins, the second defines the bands that are equally spaced between minimum and maximum value. That is, the first band from 11.0-23.0 has 19 entries, the second, from 23.0-35.0 has 102 entries, etc. 


Calculate the histogram for precipitation and temperature when b=20. Once you have the calculated bins and values, plot the histogram for precipitation and temperature using `hist` function in `matplotlib.pyplot` library. Provide a written interpretation of your results. Comment on the shape of each histogram; is it unimodal or bimodal? Is the distribution skewed (if so, in which direction)?

Explain how changing the number of bins affects the appearance of the histogram.

```python
def histogram(vectors, b):
    vector_min = min(vectors)
    vector_max = max(vectors)
    #calculate number of bins.
    number_of_bins = (vector_max - vector_min) / b
    #create bin borders
    borders = []
    for i in range(b +1):
        borders.append(vector_min + i*number_of_bins)

    #count values in each bin
    counts = []
    for i in range(b):
        counts.append(0)

    for value in vectors:
        index = int((value - vector_min) / number_of_bins)
        if index == b:  # Handle max vector
            index = b - 1
        counts[index] += 1
    return [counts,borders]


# the call to your function
print(histogram(temperature_vector, b = 5))
print(histogram(precipitation_vector, b = 5))

# Plot 1: Temperature
plt.hist(temperature_vector, bins=20, color='steelblue', edgecolor='black')
plt.xlabel('Temperature (degrees F)')
plt.ylabel('Frequency')
plt.title('Temperature Histogram (b=20)')
plt.show()

# Plot 2: Precipitation
plt.hist(precipitation_vector, bins=20, color='green', edgecolor='black')
plt.xlabel('Precipitation (inches)')
plt.ylabel('Frequency')
plt.title('Precipitation Histogram (b=20)')
plt.show()
```

### Part 3.1. Descriptive statistics

Describe what a Poisson random variable is. What is the parameter, $\lambda$? What is the min, max, mean, and variance of a Poisson random variable? 

**Your description:**

A Poisson random variable models the number of events occurring in a fixed interval of time or space, given a constant average rate of occurrence. It's used for events that occur independently and randomly.

$\lambda$(lambda) is the average rate parameter - it represents the average number of occurrences of an event in the given time or space interval. This value remains constant throughout the observed interval and determines the shape of the distribution.

Min - 0
Max - infinity
Mean - E[X] = $\lambda$
Variance - Var[X] = $\lambda$

Poisson distribution, the mean and variance are both equal to $\lambda$. This is a unique characteristic of the Poisson distribution.
